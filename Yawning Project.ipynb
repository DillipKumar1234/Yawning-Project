```python
import cv2 as cv
import numpy as np
import dlib
import time
```

```python
face_detector=dlib.get_frontal_face_detector()
```

```python
landmark_predictor=dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
counter=0
x=0
z=0
q=0
l=0
v=0
c=0
n=0
cam=cv.VideoCapture(0)
while True:
    _,frame=cam.read()
    frame=cv.flip(frame,1)
    gray=cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    faces=face_detector(gray)                              ##to creat a rectangle in face
    for face in faces:
        landmarks=landmark_predictor(gray,face)
        for point in range(47,68):
            upper_lip_x=int(landmarks.part(52).x)
            upper_lip_y=int(landmarks.part(52).y)
            
            lower_lip_x=int(landmarks.part(56).x)
            lower_lip_y=int(landmarks.part(56).y)
            
            cv.circle(frame,(upper_lip_x,upper_lip_y),3,(255,0,0),-1)
            cv.circle(frame,(lower_lip_x,lower_lip_y),3,(255,0,0),-1)
            
            
            left_lip_x=int(landmarks.part(48).x)
            left_lip_y=int(landmarks.part(48).y)
            
            right_lip_x=int(landmarks.part(54).x)
            right_lip_y=int(landmarks.part(54).y)
            cv.circle(frame,(left_lip_x,left_lip_y),3,(255,0,0),-1)
            cv.circle(frame,(right_lip_x,right_lip_y),3,(255,0,0),-1)
    
            ratio=(right_lip_x-left_lip_x)/(lower_lip_y-upper_lip_y)
        
            print(ratio)
            
            
            if(ratio<=2):
                cv.putText(frame,'yawning',(100,100),cv.FONT_HERSHEY_SIMPLEX,0.7,(255,0,0),2)
                
                z=time.localtime()
                q=time.strftime("%H:%M:%S",z)
                l=int(q[6]+q[7])
                
                
                
            else:
                
                
                if(l<x):
                    v=l-x
                if(l>x):
                    v=(60-l)+x
                if(v>3):
                    
                    counter+=1
                if(counter>3):
                    cv.putText(frame,'!Be Ware:Alert!',(310,100),cv.FONT_HERSHEY_SIMPLEX,0.7,(255,0,0),2)
                z=0
                q=0
                l=0
                
                
                
                cv.putText(frame,'active',(200,100),cv.FONT_HERSHEY_SIMPLEX,0.7,(255,0,0),2)
                n=time.localtime()
                c=time.strftime("%H:%M:%S",n)
                x=int(c[6]+c[7])
                
                
    cv.imshow("YAWNING",frame)
    key=cv.waitKey(30)
    if(key==27):
        break

cv.destroyAllWindows()
cam.release()
    
```
